```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Завантажуємо дані
df = pd.read_csv('data.csv')

# Огляд даних
print(df.head())  # Перегляд перших рядків
print(df.info())  # Інформація про колонки
print(df.describe())  # Описова статистика

# Візуалізація розподілу змінних
sns.histplot(df['Salary'])
plt.title('Розподіл змінної Salary')
plt.show()

# Фільтруємо тільки числові колонки для кореляції
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Кореляція між змінними
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Кореляція між змінними')
plt.show()

# Boxplot для виявлення аномалій
sns.boxplot(x=df['Salary'])
plt.title('Ящик з вусами для Salary')
plt.show()

# Перевірка на пропуски
print(df.isnull().sum())

# Обробка пропусків (наприклад, заповнення середнім значенням)
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
```
