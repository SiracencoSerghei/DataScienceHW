```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

---
# Visualisation
---

```python
df = pd.read_csv('MrBeast_youtube_stats.csv')
```

```python
def pretty_plot(df):
    plt.figure(figsize=(12,8))
    plt.plot(df.likeCount[50:100], 'o-',color='orange', label = 'likes')
    # plt.plot(df.viewCount[50:100], label = 'Viewes', color = 'magenta', linestyle=':')
    plt.plot(df.viewCount[50:100],'m*-', label = 'Viewes')

    plt.xlabel('Video on Youtube')
    plt.ylabel('Likes quantity', color='Blue', fontsize = 'larger')
    plt.title('Youtube video analytics', color='tomato', fontsize = 'larger')
    plt.legend(loc='upper left')
    plt.grid()
    plt.show()
```

```python
pretty_plot(df)
```

```python
def ratio(df):
    plt.figure(figsize=(12,8))
    plt.plot(df.viewCount[50:100], df.likeCount[50:100], 'mo', label='Likes(Views)')

    plt.xlabel('View Count')
    plt.ylabel('Like Count')
    plt.title('Likes vs Views from Index 50 to 100')

    plt.legend()
    plt.grid()

    plt.show() 
```

```python
ratio(df)
```
