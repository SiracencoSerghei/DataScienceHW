# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
date = pd.date_range(start = '2021-09-01', freq= 'D', periods = 8)
plt.figure(figsize=(15,6))
plt.plot(date, [23, 17, 17, 16, 15, 14, 17, 20])

plt.show()
```

### Текст на графіках
xlabel(або ylabel) - рядок тексту подпісу. А також параметри конструктора класу matplotlib.text. Text:
*fontsize* або *size* - розмір шрифту, число або значення зі списку *xx-small*, *x-small*, *small*, *medium*, *large*, *x-large*, *xx-large*
*fontstyle* - стиль шрифту значення зі списку normal, italic, oblique.
*fontweight* - товщина шрифту може визначатися або числом в діапазоні від 0 до 1000, або значенням зі списку *ultralight*, *light*, *normal*, *regular*, *book*, *medium*, *roman*, *semibold*, *demibold*, *demi*, *bold*, *heavy*, *extra bold*, *black*.
*color* - значення кольору для тексту підпису. Можливості вибору кольору занадто великі і подивитися варіанти можна за цім посиланням [(https://matplotlib.org/stable/gallery/color/named_colors.html#sphx-glr-gallery-color-named-colors-py)], де вони теж наведені не повностю.

```python
date = pd.date_range(start="2021-09-01", freq="D", periods=8)
plt.figure(figsize=(15, 6), facecolor="lightskyblue")
plt.plot(date, [23, 17, 17, 16, 15, 14, 17, 20], marker='o', linestyle='--', color='b', label = 'day temperature')
plt.xlabel('Date', fontsize='large', fontweight='bold', color='midnightblue')
plt.xticks(rotation=45, fontsize=20, fontweight='bold', color='darkblue')
plt.ylabel('Temperature (C)', fontsize='large', fontweight='bold', color='midnightblue')
plt.yticks(fontsize=20, fontweight='bold', color='darkblue')
plt.text(date[0], 15, "Осінь досить тепла", color='darkgreen', fontsize=20, fontweight='bold')
plt.title("Temperature in Poltava", fontsize=20, fontweight="bold", color="darkblue")
plt.legend(loc='upper right', fontsize=20)
plt.grid(True, linestyle='--', linewidth=0.5, color='lightgray')
plt.tight_layout()
plt.show()
```

```python
# Компонування графиків

date = pd.date_range(start="2021-09-01", freq="D", periods=8)
fig, axs = plt.subplots(figsize=(15, 6), facecolor="lightskyblue")
axs.plot(date, [23, 17, 17, 16, 15, 14, 17, 20], marker='o', linestyle='--', color='b', label='day temperature')
axs.plot(date, [19, 11, 16, 11, 10, 10, 11, 16], marker='o', linestyle='-', color='r', label='night temperature')
plt.xlabel('Date', fontsize='large', fontweight='bold', color='midnightblue')
plt.ylabel('Temperature (C)', fontsize='large', fontweight='bold', color='midnightblue')
plt.text(date[0], 15, "Осінь досить тепла", color='darkgreen', fontsize=20, fontweight='bold')
plt.title("Temperature in Poltava", fontsize=20, fontweight="bold", color="darkblue")
plt.legend(loc='upper right', fontsize=20)
plt.show()
```

```python
date = pd.date_range(start="2021-09-01", freq="D", periods=8)
fig, axs = plt.subplots(2, 1, figsize=(15, 12), facecolor="lightskyblue")

axs[0].plot(date, [23, 17, 17, 16, 15, 14, 17, 20], marker='o', linestyle='--', color='b', label='day temperature')
axs[1].plot(date, [19, 11, 16, 11, 10, 10, 11, 16], marker='o', linestyle='-', color='r', label='night temperature')

axs[0].set_title("Day Temperature in Poltava", fontsize=20, fontweight="bold", color="darkblue")
axs[1].set_title("Night Temperature in Poltava", fontsize=20, fontweight="bold", color="darkblue")

axs[0].set_ylabel('Temperature (C)', fontsize='large', fontweight='bold', color='midnightblue')
axs[1].set_ylabel('Temperature (C)', fontsize='large', fontweight='bold', color='midnightblue')
axs[1].set_xlabel('Date', fontsize='large', fontweight='bold', color='midnightblue')
axs[0].set_xlabel('Date', fontsize='large', fontweight='bold', color='midnightblue')

fig.suptitle("Temperature in Poltava", fontsize=20, fontweight="bold", color="darkblue")

# fig.tight_layout(pad = 3.0)
plt.show()
```

```python
# Стовпчикови діаграми

data = pd.DataFrame(
    {
        "City": ["Kyiv", "Lviv", "Odesa", "Kharkiv"],
        "Population": [2962180, 721301, 1012175, 1446000],
        "Color": ["blue", "green", "red", "purple"],
        "Hatch": ["/", "\\", "|", "-"],
    }
)

# Побудова діаграми
plt.figure(figsize=(10, 6), facecolor="lightskyblue")
bars = plt.bar(data["City"], data["Population"], color=data["Color"], edgecolor="black")

# Додати штрихування (hatch)
for bar, hatch in zip(bars, data["Hatch"]):
    bar.set_hatch(hatch)

# Заголовок і підписи
plt.title("Населення міст України")
plt.ylabel("Населення")
plt.xlabel("Місто")

# Показати графік
plt.show()
```

```python
# Кругова діаграма

data = pd.DataFrame(
    {
        "City": ["Kyiv", "Lviv", "Odesa", "Kharkiv"],
        "Population": [2962180, 721301, 1012175, 1446000],
        "Color": ["blue", "green", "red", "purple"],
    }
)
# Побудова кругової діаграми
plt.figure(figsize=(10, 6), facecolor="lightskyblue")
plt.pie(
    data["Population"],
    labels=data["City"],
    colors=data["Color"],
    autopct="%.2f%%",
    startangle=50,
    explode=(0.1, 0, 0, 0),  # Відокремити перший сектор
)
plt.title("Розподіл населення міст України")
plt.axis("equal")  # Зробити кругову діаграму круглою
plt.show()
```

```python
# діаграми в полярних координатах
theta = np.linspace(0, 2 * np.pi, 100)
r = np.abs(np.sin(2 * theta) * np.cos(2 * theta))
plt.polar(theta, r, color='blue', linewidth=2)
plt.title("Polar Plot Example", fontsize=20, fontweight="bold", color="darkblue")
plt.show()
```

```python
# тривимірні графіки
from mpl_toolkits.mplot3d import Axes3D

# Створення сітки
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Побудова графіка
fig = plt.figure(figsize=(14, 10), facecolor="lightskyblue", dpi=100)
ax = fig.add_subplot(111, projection="3d")
surf = ax.plot_surface(X, Y, Z, cmap="viridis")

# Оформлення
ax.set_title("3D Surface Plot", fontsize=20, fontweight="bold", color="darkblue")
ax.set_xlabel("X axis", fontsize=14, fontweight="bold", color="midnightblue")
ax.set_ylabel("Y axis", fontsize=14, fontweight="bold", color="midnightblue")
ax.set_zlabel("Z axis", fontsize=14, fontweight="bold", color="midnightblue")

plt.show()
```

```python
# діаграма розсіювання

# Генерація випадкових даних
np.random.seed(0)
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)
colors = z  # Колір залежить від z
sizes = z * 100  # Розмір залежить від z

# Створення 3D графіку
fig = plt.figure(figsize=(10, 6), facecolor="lightskyblue")
ax = fig.add_subplot(111, projection="3d")

# Побудова діаграми розсіювання
scatter = ax.scatter(
    x, y, z, c=colors, s=sizes, cmap="viridis", alpha=0.7, edgecolors="black"
)

# Додаткове оформлення
ax.set_title(
    "3D Діаграма розсіювання", fontsize=18, fontweight="bold", color="darkblue"
)
ax.set_xlabel("Вісь X", fontsize=12, fontweight="bold")
ax.set_ylabel("Вісь Y", fontsize=12, fontweight="bold")
ax.set_zlabel("Вісь Z", fontsize=12, fontweight="bold")

# Колірна шкала
fig.colorbar(scatter, ax=ax, shrink=0.6, aspect=10, label="Z (інтенсивність)")

plt.show()
```

# SEABORN

```python
import seaborn as sns
```

```python
# data from https://github.com/mwaskom/seaborn-data/blob/master/mpg.csv
# набір mpg містить дані про автомобілі, їх характеристики та витрату пального
data = sns.load_dataset('mpg')
```

```python
data
```

```python
# стиль оформлення графіків
# глобальне налаштування стилю
sns.set_style("whitegrid")
```

```python
date = pd.date_range(start="2021-09-01", freq="D", periods=8)
day = [23, 17, 17, 16, 15, 14, 17, 20]
night = [19, 11, 16, 11, 10, 10, 11, 16]
df = pd.DataFrame({"date": date, "day temperature": day, "night temperature": night})
df.head()
```

```python
# функція lineplot
sns.lineplot(data=df)
```

```python
# графік залежності нічної температури від денної
# світліша областьнавколо лінії - це 95% довірчий інтервал
sns.lineplot(x='night temperature', y='day temperature', data=df)
```

```python
# за замовчуванням seaborn  здійснює сортування набору даних
sns.lineplot(x='night temperature', y='day temperature', data=df, sort=False)
```

```python
# графік залежності потужності автомобіля від року випуску окремо для кожного регіону
sns.set_style("darkgrid")
data = sns.load_dataset('mpg')
sns.lineplot(x='model_year', y='horsepower', hue='origin', data=data, palette='Set1')
```

```python
# точковий графік залежності дальності пробігу (displacement) автомобіля від витрати пального (mpg)
sns.scatterplot(x='mpg', y='displacement', data=data)
```

```python
# можемо виділлити кольором регіон виробника
sns.scatterplot(x='mpg', y='displacement', hue='origin', data=data)
```

```python
# ця функція надає можливість налаштовувати не лише зовнішній вигляд графіка, алє й фігуру подкладку, на якій розміщені всі графічни компоненти
sns.relplot(x='mpg', y='displacement', kind='scatter', data=data)
```

```python
# діаграму можна розмістити на трьох окремих полях, як роздільний параметер будемо використовувати регіон виробника (origin)
sns.relplot(x='mpg', y='displacement', kind='scatter', data=data, col='origin', hue='origin')
```

```python
# Візуалізація категоріальних даних
# вониприймають значення з невпорядкованоЇ множини. Це ознакі кольору, форми, розміру тощо.
# до них не можна застосовувати арифметичні операції, операцію порівняння, але можна використовувати для групування даних
sns.stripplot(x='origin', y='mpg', data=data)
```

```python
# виділимо кольором кількість циліндрів у двігуна
sns.stripplot(x='origin', y='mpg', hue='cylinders', data=data, palette='Set1')
```

```python
# точкі, що відображаються на діаграмі, не перекриваються, це дозволяє робити висновки про переважання тих чи інших значень в наборі даних
sns.swarmplot(x='origin', y='mpg', data=data)
```

```python
# візуалізаціЇ розподілу категоріальних даних

# будує діаграму типу "скринька з вусами" (boxplot), на ній відображаються медіанне значення, квартилі та викиди
sns.boxplot(x='origin', y='mpg', data=data, hue='cylinders', palette='Set1')
```

```python
# density plot
sns.violinplot(x='origin', y='mpg', data=data)
```

```python
# будує діаграму з прямокутників, добре підходять для візуалізаціЇ великіх наборів даних
sns.boxenplot(x='origin', y='mpg', data=data, hue='cylinders', palette='Set1')
```

```python
# відображає набір даних та лінію регресії, побудовану за ними.
# лінію регресії залежности відстані, яку може проїхати автомобіль на одному галоні пального (mpg) від його потужності (horsepower)
sns.lmplot(x='horsepower', y='displacement', data=data)
```

```python
sns.lmplot(x='horsepower', y='displacement', data=data, hue='origin', palette='Set1', col='origin')
```
