## Матричні розклади

**Матричні розклади** - це метод аналізу даних, що використовується для зменшення розмірності даних та виявлення прихованих залежностей між об'єктами або користувачами та їх властивостями. У контексті рекомендаційних систем, матричний розклад допомагає моделювати взаємодії між користувачами і об'єктами (такими як фільми, музичні треки, товари тощо) за допомогою розкладання великої матриці взаємодій на декілька менших матриць.

Основні методи матричного розкладу, які використовуються в рекомендаційних системах:
1. Singular Value Decomposition (SVD)
2. Non-negative Matrix Factorization (NMF)
3. Factorization Machines (FM)

Математична особливість методу Matrix Factorization полягає в розкладанні матриці рейтингів користувачів та предметів на добуток двох або більше матриць меншого розміру.

Для матриці рейтингів $R$ розміром $m×n$, де $m$ - кількість користувачів, а $n$ - кількість предметів (або ознак), Matrix Factorization зазвичай розкладає цю матрицю у вигляді добутку двох матриць меншого розміру $P$ та $Q$.

Математично, це можна виразити так:
$$R≈P×Q^T$$

Де:

$R$ - матриця рейтингів користувачів та предметів розміром $m×n$.

$P$ - матриця факторів користувачів розміром $m×k$, де $k$ - кількість факторів.

$Q$ - матриця факторів предметів розміром $n×k$.

Мета полягає в тому, щоб знайти такі матриці $P$ та $Q$, які найкраще апроксимують вихідну матрицю $R$. Це дозволяє нам використовувати відомості про спільні характеристики користувачів та предметів для передбачення невідомих рейтингів і рекомендацій

## Singular Value Decomposition (SVD)

![alt text](https://upload.wikimedia.org/wikipedia/commons/thumb/b/bb/Singular-Value-Decomposition.svg/1200px-Singular-Value-Decomposition.svg.png)

SVD - це один з найпоширеніших методів матричного розкладу. Він полягає у розкладанні початкової матриці взаємодій на три менші матриці: **U**, **Sigma** і **V**.
- Матриця U відображає взаємодії користувачів,
- матриця V - взаємодії об'єктів,
- а матриця Sigma - діагональна матриця, яка містить сингулярні значення.

SVD може бути застосований як до повної матриці взаємодій, так і до розріджених матриць (тобто матриць з великою кількістю відсутніх значень).

Приклад: **матриця переваг фільму**

Уявімо, що у нас є гігантська матриця, яка показує, наскільки кожній людині подобається кожен фільм. Кожен рядок у матриці представляє людину, кожен стовпець представляє фільм, а числа в матриці представляють оцінки, які люди дають фільмам.

![alt text](https://miro.medium.com/v2/resize:fit:1400/0*aWs9lrTAgkC9fRXh)
Джерело: https://medium.com/explorium-ai/building-a-recommender-system-for-data-enrichment-b289a1a054c5


Ряди (n) (Люди): У різних людей різні смаки. Комусь більше подобаються бойовики, а комусь більше романтичні фільми.

Колонки (m) (фільми): фільми також мають свої унікальні характеристики. Деякі з них насичені діями, деякі смішні, а деякі емоційні.

Тепер SVD допомагає нам розбити цю гігантську матрицю на три менші матриці:

1. **Матриця користувачів (U):** Ця матриця показує, наскільки кожній людині подобається певний тип фільму. Кожен рядок у цій матриці говорить нам про загальний смак людини до фільмів. Значення в цій матриці допомагають охопити унікальні вподобання кожної людини.

2. **Матриця предметів (V):** Ця матриця представляє характеристики кожного фільму. Кожен стовпець у цій матриці розповідає нам про певний аспект фільму, наприклад, скільки в ньому дії або наскільки він смішний. Значення в цій матриці допомагають відобразити унікальні особливості кожного фільму.

3. **Діагональна матриця (Sigma):** ця матриця показує, наскільки важливим є кожен тип уподобання фільму. Це допомагає нам визначати, які аспекти фільму (наприклад, бойовик, романтика) є більш важливими для визначення рекомендацій.

Магія відбувається, коли ми знову множимо ці матриці (U * Sigma * V). Ця матриця продуктів дає нам приблизні оцінки для кожної пари користувач-фільм. Це схоже на поєднання загальних уподобань людей (U), унікальних особливостей кожного фільму (V) і важливості різних аспектів (Sigma), щоб передбачити, наскільки комусь може сподобатися фільм, якого вони не бачили.

```python
import numpy as np
import pandas as pd

# Sample dataset (user_id, item_id, rating)
data = {'user_id': [1, 1, 2, 2, 3, 3, 4, 4],
        'item_id': ['A', 'B', 'A', 'C', 'C', 'D', 'B', 'D'],
        'rating': [5, 4, 3, 2, 4, 5, 1, 3]}

df_sample = pd.DataFrame(data)

# Convert records into user-book review score matrix
datamatrix = df_sample.pivot(index="user_id", columns="item_id", values="rating").fillna(0)
matrix = datamatrix.values
datamatrix
```

```python
matrix
```

```python
# Singular value decomposition
u, s, vh = np.linalg.svd(matrix, full_matrices=False)

# Find the highest similarity
def cosine_similarity(v,u):
    return (v @ u)/ (np.linalg.norm(v) * np.linalg.norm(u))
```

```python
vh
```

```python
vh[:,0]
```

```python
target_item = 0  # A

highest_similarity = -np.inf
highest_sim_col = -1
for col in range(1,vh.shape[1]):
    similarity = cosine_similarity(vh[:,target_item], vh[:,col])
    if similarity > highest_similarity:
        highest_similarity = similarity
        highest_sim_col = col

print(f"Item {highest_sim_col} (item_id {datamatrix.columns[col]}) is most similar to item {target_item} (item_id {datamatrix.columns[target_item]})")
```

## Різниця між SVD та SVD++


У бібліотеці Surprise (яка використовується для рекомендаційних систем), SVD та SVD++ є двома різними алгоритмами для прогнозування рейтингів. Ось їх основні відмінності:

- SVD (Singular Value Decomposition):

Це базова модель, яка використовує метод зменшення розмірності матриці рейтингів.
Модель використовує тільки інформацію про рейтинги користувачів для прогнозування нових рейтингів.

$$\hat{r}_{ui} = \mu + b_u + b_i + q_i^T \cdot p_u $$


- SVD++ (SVD with Implicit Feedback):

Це розширення моделі SVD, яке також враховує неявний фідбек (implicit feedback), такий як інформація про те, як користувачі взаємодіють з об'єктами (наприклад, час, який вони проводять переглядаючи сторінку, кількість кліків тощо). SVD++ включає інформацію про користувачів та об'єкти в процес прогнозування рейтингів, що може покращити точність рекомендацій.

$$\hat{r}_{ui} = \mu + b_u + b_i + q_i^T \cdot \left( p_u + \frac{1}{\sqrt{|N(u)|}} \cdot \sum_{j \in N(u)} y_j \right) $$

$|N(u)|$ - кількість об'єктів, з якими взаємодіяв користувач $u$;

$y_j$ - додатковий вектор для об'єкта j, що представляє неявний фідбек користувачів з об'єктом j.(вектор факторів для товару j).

```python
from surprise import Dataset, Reader
from surprise.model_selection import train_test_split
from surprise import SVD, SVDpp
from surprise import accuracy

data = Dataset.load_builtin('ml-100k')
reader = Reader(rating_scale=(1, 5))
trainset, testset = train_test_split(data, test_size=0.2)
```

```python
svd_model = SVD(n_factors=100,
                n_epochs=20,
                lr_all=0.005,
                reg_all=0.02)
svd_model.fit(trainset)
predictions_svd = svd_model.test(testset)
accuracy_svd = accuracy.rmse(predictions_svd)
accuracy_svd
```

```python
svdpp_model = SVDpp(n_factors=100,
                n_epochs=20,
                lr_all=0.005,
                reg_all=0.02)
svdpp_model.fit(trainset)
predictions_svdpp = svdpp_model.test(testset)
accuracy_svdpp = accuracy.rmse(predictions_svdpp)
accuracy_svdpp
```
